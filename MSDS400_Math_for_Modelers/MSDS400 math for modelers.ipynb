```python
import numpy as np

# Coefficient matrix (3x3)
A = np.array([[1, -3, -4],
              [2, 6, 3],
              [-6, 4, 5]])

# Constants vector
b = np.array([-1, 0, 16])

# Compute determinant of A
D = np.linalg.det(A)

if np.isclose(D, 0):
    print("System has no unique solution (D = 0)")
else:
    Dx = A.copy()
    Dx[:, 0] = b
    Dx_det = np.linalg.det(Dx)

    Dy = A.copy()
    Dy[:, 1] = b
    Dy_det = np.linalg.det(Dy)

    Dz = A.copy()
    Dz[:, 2] = b
    Dz_det = np.linalg.det(Dz)

    x = Dx_det / D
    y = Dy_det / D
    z = Dz_det / D

    print(f"D  = {D:.2f}")
    print(f"Dx = {Dx_det:.2f}")
    print(f"Dy = {Dy_det:.2f}")
    print(f"Dz = {Dz_det:.2f}\n")

    print("Solution using Cramer's Rule:")
    print(f"x = {x:.4f}")
    print(f"y = {y:.4f}")
    print(f"z = {z:.4f}")


if D == 0:
    print("System has no unique solution (determinant is zero)")

```

Cofactor C1,2

```python
import numpy as np

# Define a 3x3 matrix
A = np.array([[5, -1, 8],
              [3, 0, -1],
              [1, 3, 6]])

# Coordinates for cofactor C[1,2] (Python is 0-indexed, so it's row 0, col 1)
i, j = 0, 1

# Get the minor by deleting row i and column j
minor = np.delete(np.delete(A, i, axis=0), j, axis=1)

# Calculate determinant of the minor
minor_det = np.linalg.det(minor)

# Calculate the cofactor
cofactor = ((-1) ** (i + j)) * minor_det

print(f"Cofactor C[1,2] = {cofactor:.2f}")
```

Cofactor Expansion along row 4 of 4x4

```python
from sympy import Matrix

# Create the matrix with specific numerical values
A = Matrix([
    [-4, 3, -4, 4],
    [5, -2, -1, 2],
    [0, 3, 0, 1],
    [-4, 1, 4, -4]
])

# Extract the values from row 4
w = A[3, 0]  # a41 = 0
x = A[3, 1]  # a42 = 4
y = A[3, 2]  # a43 = 2
z = A[3, 3]  # a44 = 3

# Get the cofactor matrices (the submatrices)
minor_A41 = A.minor_submatrix(3, 0)
minor_A42 = A.minor_submatrix(3, 1)
minor_A43 = A.minor_submatrix(3, 2)
minor_A44 = A.minor_submatrix(3, 3)

# Print the cofactor matrices
print("Cofactor Matrix A41:")
print(minor_A41)
print("\nCofactor Matrix A42:")
print(minor_A42)
print("\nCofactor Matrix A43:")
print(minor_A43)
print("\nCofactor Matrix A44:")
print(minor_A44)

# Calculate the determinants of the cofactor matrices
A41 = minor_A41.det()
A42 = minor_A42.det()
A43 = minor_A43.det()
A44 = minor_A44.det()

# Apply the cofactor signs
A41_signed = (-1)**(3+0) * A41  # (-1)^(4+1) * A41
A42_signed = (-1)**(3+1) * A42  # (-1)^(4+2) * A42
A43_signed = (-1)**(3+2) * A43  # (-1)^(4+3) * A43
A44_signed = (-1)**(3+3) * A44  # (-1)^(4+4) * A44

# Calculate the final determinant
det = w * A41_signed + x * A42_signed + y * A43_signed + z * A44_signed

# Print the determinants of the cofactor matrices
print("\nDeterminants of the cofactor matrices:")
print(f"det(A41) = {A41}")
print(f"det(A42) = {A42}")
print(f"det(A43) = {A43}")
print(f"det(A44) = {A44}")

# Print the formula and values
print("\nFinal calculation:")
print(f"det(A) = {w} * ({A41_signed}) + {x} * ({A42_signed}) + {y} * ({A43_signed}) + {z} * ({A44_signed})")
print(f"det(A) = {w*A41_signed} + {x*A42_signed} + {y*A43_signed} + {z*A44_signed}")
print(f"det(A) = {det}")

# Verification
print("\nVerification using SymPy's determinant function:")
print(f"det(A) = {A.det()}")
```

```python
# Given information
det_RQ = 25
det_QC = 13
det_CH = 12

# Calculate detRQ^T
# The determinant of a transpose is equal to the determinant of the original matrix
det_RQ_transpose = det_RQ
print(f"det(RQ^T) = det(RQ) = {det_RQ_transpose}")

# Calculate detRC^-1
# For any invertible matrix A, det(A^-1) = 1/det(A)
# We know R(C^-1), but we need det(C)
# Since det(QC) = det(Q) * det(C), we need det(Q)
# From det(RQ) = det(R) * det(Q), we can isolate det(Q)
# However, we don't have det(R) directly

# Let's use what we have:
# If we assume RC^-1 is meant as R * (C^-1), then:
# det(R * C^-1) = det(R) * det(C^-1) = det(R) * (1/det(C))

# Since we don't have det(R) or det(C) directly, we need to derive them:
# From RQ, QC, and CH relationships
# We know det(RQ) = det(R) * det(Q) = 25
# We know det(QC) = det(Q) * det(C) = 13
# We know det(CH) = det(C) * det(H) = 12

# From det(QC) = 13, we can find:
# det(C) = 13 / det(Q)

# From det(RQ) = 25, we can find:
# det(Q) = 25 / det(R)

# Substituting:
# det(C) = 13 / (25 / det(R)) = (13 * det(R)) / 25

# Therefore:
# det(RC^-1) = det(R) * (1/det(C)) = det(R) * (1/((13 * det(R)) / 25)) = 25/13

det_RC_inverse = 25/13
print(f"det(RC^-1) = {det_RC_inverse}")

# Calculate detRQCH
# For matrix multiplication: det(AB) = det(A) * det(B)
# So det(RQCH) = det(R) * det(Q) * det(C) * det(H)
# We know det(RQ) = 25, det(QC) = 13, det(CH) = 12

# det(RQCH) = det(RQ) * det(C) * det(H) / det(Q)
# Or equivalently:
# det(RQCH) = det(R) * det(QC) * det(H)
# Or:
# det(RQCH) = det(RQ) * det(CH) / det(C)
#
# Since we don't have individual determinants, only products, let's use:
# det(RQCH) = det(RQ) * det(CH) / det(C) * det(Q)
#
# From above, det(Q) = 25 / det(R) and det(C) = 13 / det(Q) = 13 * det(R) / 25
#
# Substituting into det(RQCH) = det(R) * det(Q) * det(C) * det(H):
# det(RQCH) = det(R) * (25/det(R)) * ((13*det(R))/25) * (12/det(C))
# det(RQCH) = 25 * 13 * 12 / 25 = 13 * 12 = 156

det_RQCH = 25 * 12
print(f"det(RQCH) = {det_RQCH}")

# Calculate detRH
# We don't have det(R) or det(H) directly, nor do we have a direct relation between R and H
# We know:
# det(RQ) = 25 = det(R) * det(Q)
# det(CH) = 12 = det(C) * det(H)
# det(QC) = 13 = det(Q) * det(C)
#
# From these, we can derive:
# det(H) = 12 / det(C)
# det(C) = 13 / det(Q)
# det(Q) = 25 / det(R)
#
# Substituting:
# det(H) = 12 / (13 / det(Q)) = 12 * det(Q) / 13
# det(H) = 12 * (25 / det(R)) / 13 = 12 * 25 / (13 * det(R))
#
# Therefore:
# det(RH) = det(R) * det(H) = det(R) * (12 * 25 / (13 * det(R))) = 12 * 25 / 13

det_RH = 12 * 25 / 13
print(f"det(RH) = {det_RH}")
```

```python
from scipy.optimize import linprog

# Objective function: Revenue = 20m + 19r
# m = 4p + 2e + 15f
# r = 2p + 4e + 17f
# So, Revenue = 20*(4p + 2e + 15f) + 19*(2p + 4e + 17f)
# = 118p + 116e + 623f

c = [-118, -116, -623]  # Coefficients (negated for maximization)

# Constraints: p ≤ 460, e ≤ 420, f ≤ 1915
A = [
    [1, 0, 0],  # p constraint
    [0, 1, 0],  # e constraint
    [0, 0, 1]   # f constraint
]
b = [460, 420, 1915]

# Variable bounds: p, e, f ≥ 0
bounds = [(0, None), (0, None), (0, None)]

# Solve using linear programming
result = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

# Output results
if result.success:
    p, e, f = result.x
    revenue = -result.fun  # Convert back to positive
    print(f"Optimal p: {p:.0f}")
    print(f"Optimal e: {e:.0f}")
    print(f"Optimal f: {f:.0f}")
    print(f"Maximum Revenue: {revenue:.2f}")
else:
    print("Optimization failed.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog

def solve_with_scipy():
    """
    Maximize p = 9x + 7y
    Subject to:
    1. 20x + 19y <= 41
    2. 17x + 5y <= 54
    3. x >= 0
    4. y >= 0
    """
    # Since linprog minimizes by default, we negate the objective coefficients
    c = np.array([-9, -7])  # Negated for maximization

    # Constraint coefficients
    A = np.array([
        [20, 19],  # 20x + 19y <= 41
        [17, 5]    # 17x + 5y <= 54
    ])

    # Constraint right-hand sides
    b = np.array([41, 54])

    # Bounds for variables (x >= 0, y >= 0)
    bounds = [(0, None), (0, None)]

    # Solve the linear programming problem
    result = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

    # Extract the solution
    if result.success:
        x_opt, y_opt = result.x
        p_max = -result.fun  # Negate back to get the maximum

        print("Optimal solution from scipy:")
        print(f"x = {x_opt}")
        print(f"y = {y_opt}")
        print(f"Maximum value of p = 9x + 7y = {p_max}")

        # Verify the constraints
        print("\nVerifying constraints:")
        print(f"20x + 19y = 20({x_opt}) + 19({y_opt}) = {20*x_opt + 19*y_opt} <= 41")
        print(f"17x + 5y = 17({x_opt}) + 5({y_opt}) = {17*x_opt + 5*y_opt} <= 54")

        return {
            "x": x_opt,
            "y": y_opt,
            "p_max": p_max
        }
    else:
        print("Optimization failed:", result.message)
        return None

def check_all_corner_points():
    """
    Check all corner points of the feasible region to find the optimal solution
    """
    print("\nChecking all corner points:")

    # Corner points of the feasible region
    corners = []

    # Corner at (0, 0)
    corners.append((0, 0))

    # Corner at intersection of 20x + 19y = 41 and x = 0
    # 19y = 41 => y = 41/19
    corners.append((0, 41/19))

    # Corner at intersection of 17x + 5y = 54 and y = 0
    # 17x = 54 => x = 54/17
    corners.append((54/17, 0))

    # Corner at intersection of 20x + 19y = 41 and 17x + 5y = 54
    # Let's solve the system of equations
    A = np.array([[20, 19], [17, 5]])
    b = np.array([41, 54])

    try:
        solution = np.linalg.solve(A, b)
        x_int, y_int = solution

        # Check if this point satisfies all constraints
        if x_int >= 0 and y_int >= 0:
            corners.append((x_int, y_int))
    except np.linalg.LinAlgError:
        print("The constraints are parallel or dependent.")

    # Evaluate the objective function at each corner point
    print("Corner points and their objective values:")
    print("Point (x, y) | 20x + 19y | 17x + 5y | p = 9x + 7y")
    print("-------------|-----------|----------|------------")

    max_p = float('-inf')
    optimal_corner = None

    for x, y in corners:
        p = 9*x + 7*y
        constraint1 = 20*x + 19*y
        constraint2 = 17*x + 5*y

        print(f"({x:.4f}, {y:.4f}) | {constraint1:.4f} | {constraint2:.4f} | {p:.4f}")

        # Check if this is the optimal corner
        if p > max_p:
            max_p = p
            optimal_corner = (x, y)

    print(f"\nOptimal corner point: ({optimal_corner[0]:.4f}, {optimal_corner[1]:.4f})")
    print(f"Maximum value of p = 9x + 7y = {max_p:.4f}")

    return optimal_corner, max_p

def graphical_solution():
    """
    Provide a graphical solution to visualize the feasible region and optimal point
    """
    # Create a grid of points
    x = np.linspace(0, 5, 1000)
    y = np.linspace(0, 3, 1000)
    X, Y = np.meshgrid(x, y)

    # Calculate the objective function values
    P = 9*X + 7*Y

    # Define the constraint boundaries
    y1 = (41 - 20*x) / 19  # 20x + 19y = 41
    y2 = (54 - 17*x) / 5   # 17x + 5y = 54

    # Create the plot
    plt.figure(figsize=(10, 8))

    # Plot the feasible region
    valid_y1 = np.maximum(0, np.minimum(y1, 3))  # Clip to avoid negative values
    valid_y2 = np.maximum(0, np.minimum(y2, 3))  # Clip to avoid negative values

    # Find the minimum of the two constraints at each x
    min_y = np.minimum(valid_y1, valid_y2)

    # Plot the feasible region
    plt.fill_between(x, 0, min_y, where=(x >= 0) & (min_y >= 0), alpha=0.3, color='blue', label='Feasible Region')

    # Plot the constraint boundaries
    plt.plot(x, y1, 'r-', label='20x + 19y = 41')
    plt.plot(x, y2, 'g-', label='17x + 5y = 54')
    plt.axhline(y=0, color='k', linestyle='-', alpha=0.3, label='y = 0')
    plt.axvline(x=0, color='k', linestyle='-', alpha=0.3, label='x = 0')

    # Get the optimal corner point
    optimal_corner, max_p = check_all_corner_points()

    # Plot the optimal point
    plt.plot(optimal_corner[0], optimal_corner[1], 'ro', markersize=10,
             label=f'Optimal: ({optimal_corner[0]:.4f}, {optimal_corner[1]:.4f})')

    # Plot level curves of the objective function
    p_values = [5, 10, 15, 20, 25, 30]
    for p in p_values:
        y_level = (p - 9*x) / 7
        plt.plot(x, y_level, 'k--', alpha=0.5)
        plt.text(4, (p - 9*4) / 7, f'p = {p}', fontsize=8)

    # Set the plot limits
    plt.xlim(0, 5)
    plt.ylim(0, 3)

    # Add labels and title
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Maximization Problem: p = 9x + 7y')
    plt.grid(True)
    plt.legend()

    # Save the plot
    plt.savefig('feasible_region.png')

    return optimal_corner, max_p

# Solve the problem using scipy
scipy_solution = solve_with_scipy()

# Solve the problem by checking all corner points
print("\nSolving by checking all corner points:")
corner_solution, corner_max_p = check_all_corner_points()

# Solve the problem graphically
print("\nGraphical solution:")
graphical_solution()

# Explain the solution
print("\nExplanation of the solution:")
print("1. The feasible region is bounded by the constraints:")
print("   - 20x + 19y <= 41")
print("   - 17x + 5y <= 54")
print("   - x >= 0")
print("   - y >= 0")
print("2. In linear programming, the optimal solution always occurs at a corner point of the feasible region.")
print("3. We checked all corner points and found that the optimal solution is at:")
print(f"   x = {corner_solution[0]:.4f}, y = {corner_solution[1]:.4f}")
print(f"4. This gives a maximum value of p = 9x + 7y = {corner_max_p:.4f}")
print("5. The optimal point is at the intersection of the constraint 17x + 5y = 54 and y = 0.")
print("   This means that the optimal solution allocates all resources to x and none to y.")

# Verify which constraints are binding at the optimal solution
if scipy_solution:
    x_opt, y_opt = scipy_solution['x'], scipy_solution['y']
    constraint1 = 20*x_opt + 19*y_opt
    constraint2 = 17*x_opt + 5*y_opt

    print("\nBinding constraints at the optimal solution:")
    print(f"Constraint 1: 20x + 19y = {constraint1:.4f} {'≈' if abs(constraint1 - 41) < 1e-6 else '<'} 41")
    print(f"Constraint 2: 17x + 5y = {constraint2:.4f} {'≈' if abs(constraint2 - 54) < 1e-6 else '<'} 54")
    print(f"Constraint x >= 0: x = {x_opt:.4f} > 0")
    print(f"Constraint y >= 0: y = {y_opt:.4f} {'=' if abs(y_opt) < 1e-6 else '>'} 0")
```

```python
import numpy as np
from scipy.optimize import linprog

def solve_with_scipy():
    """
    Solve the linear programming problem using scipy.optimize.linprog

    Maximize p = n*6950 + r*7800
    Subject to:
    1. n*190 + r*310 <= 11370
    2. n <= 24
    3. r <= 31
    4. n, r >= 0 (implicit)
    """
    # Define the objective function coefficients (negated for minimization)
    # linprog minimizes, so we negate to maximize
    c = [-6950, -7800]

    # Define the inequality constraints (Ax <= b)
    A = [
        [190, 310],    # First constraint: n*190 + r*310 <= 11370
        [1, 0],        # Second constraint: n <= 24
        [0, 1]         # Third constraint: r <= 31
    ]

    b = [11370, 24, 31]

    # Define bounds for variables
    bounds = [(0, None), (0, None)]  # n >= 0, r >= 0

    # Solve the linear programming problem
    result = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

    # Extract the solution
    n_opt, r_opt = result.x

    # Calculate the objective function value
    max_value = -result.fun  # Negate back to get the maximum

    return {
        "optimal_point": (n_opt, r_opt),
        "max_value": max_value,
        "scipy_result": result
    }

def find_integer_solution(solution):
    """
    Find the best integer solution near the optimal point
    """
    n_opt, r_opt = solution["optimal_point"]

    # Generate candidate integer points around the optimal solution
    candidates = [
        (int(np.floor(n_opt)), int(np.floor(r_opt))),
        (int(np.floor(n_opt)), int(np.ceil(r_opt))),
        (int(np.ceil(n_opt)), int(np.floor(r_opt))),
        (int(np.ceil(n_opt)), int(np.ceil(r_opt)))
    ]

    # Filter valid candidates
    valid_candidates = []
    for n, r in candidates:
        # Check constraints
        if (n <= 24 and r <= 31 and n >= 0 and r >= 0 and
            n*190 + r*310 <= 11370):
            valid_candidates.append((n, r))

    # Find the best valid candidate
    best_candidate = None
    best_value = float('-inf')

    for n, r in valid_candidates:
        value = n*6950 + r*7800
        if value > best_value:
            best_value = value
            best_candidate = (n, r)

    return {
        "optimal_integer_point": best_candidate,
        "max_integer_value": best_value
    }

def check_corner_points():
    """
    Directly check all corner points of the feasible region
    """
    # Find the intersection of n*190 + r*310 = 11370 with the axes
    # When r = 0: n = 11370/190 = 59.84...
    # When n = 0: r = 11370/310 = 36.68...

    # But we have bounds n <= 24 and r <= 31
    # So we need to find the intersections with these bounds

    # Find r when n = 24
    # 24*190 + r*310 = 11370
    # r = (11370 - 24*190)/310 = (11370 - 4560)/310 = 6810/310 = 21.97...

    # Find n when r = 31
    # n*190 + 31*310 = 11370
    # n = (11370 - 31*310)/190 = (11370 - 9610)/190 = 1760/190 = 9.26...

    # Corner points of the feasible region
    corner_points = [
        (0, 0),                         # Origin
        (0, min(36.68, 31)),            # Intersection with r-axis, bounded by r <= 31
        (min(59.84, 24), 0),            # Intersection with n-axis, bounded by n <= 24
        (24, min(21.97, 31)),           # Intersection of n=24 with constraint
        (min(9.26, 24), 31)             # Intersection of r=31 with constraint
    ]

    # Filter valid corner points and calculate objective value
    valid_points = []
    for n, r in corner_points:
        # Round down to ensure we satisfy the constraints
        n_int = int(np.floor(n))
        r_int = int(np.floor(r))

        # Check if the point satisfies all constraints
        if (n_int <= 24 and r_int <= 31 and n_int >= 0 and r_int >= 0 and
            n_int*190 + r_int*310 <= 11370):
            valid_points.append((n_int, r_int))

    # Find the point with maximum objective value
    max_value = float('-inf')
    optimal_point = None

    for n, r in valid_points:
        value = n*6950 + r*7800
        if value > max_value:
            max_value = value
            optimal_point = (n, r)

    return {
        "optimal_point": optimal_point,
        "max_value": max_value
    }

def verify_solution(n, r):
    """
    Verify if a given solution satisfies all constraints
    """
    # Check constraints
    constraint1 = n*190 + r*310 <= 11370
    constraint2 = n <= 24
    constraint3 = r <= 31
    non_negative = n >= 0 and r >= 0

    all_satisfied = constraint1 and constraint2 and constraint3 and non_negative

    # Calculate objective value
    p = n*6950 + r*7800

    return {
        "constraint1": constraint1,
        "constraint2": constraint2,
        "constraint3": constraint3,
        "non_negative": non_negative,
        "all_satisfied": all_satisfied,
        "objective_value": p
    }

# Solve using scipy
print("SciPy solution:")
scipy_solution = solve_with_scipy()
print(f"Optimal solution: n = {scipy_solution['optimal_point'][0]}, r = {scipy_solution['optimal_point'][1]}")
print(f"Maximum value of p = {scipy_solution['max_value']}")

# Find the best integer solution
integer_solution = find_integer_solution(scipy_solution)
print("\nBest integer solution:")
print(f"n = {integer_solution['optimal_integer_point'][0]}, r = {integer_solution['optimal_integer_point'][1]}")
print(f"Maximum integer value of p = {integer_solution['max_integer_value']}")

# Check corner points directly
print("\nCorner points solution:")
corner_solution = check_corner_points()
print(f"Optimal corner point: n = {corner_solution['optimal_point'][0]}, r = {corner_solution['optimal_point'][1]}")
print(f"Maximum value at corner: p = {corner_solution['max_value']}")

# Verify the expected solution: n=24, r=21
print("\nVerifying expected solution: n=24, r=21")
verification = verify_solution(24, 21)
print(f"Constraint 1 (n*190 + r*310 <= 11370): {verification['constraint1']}")
print(f"Constraint 2 (n <= 24): {verification['constraint2']}")
print(f"Constraint 3 (r <= 31): {verification['constraint3']}")
print(f"Non-negative constraints: {verification['non_negative']}")
print(f"All constraints satisfied: {verification['all_satisfied']}")
print(f"Objective value: p = {verification['objective_value']}")

# Check if we can improve the solution
print("\nChecking if we can improve the solution:")
# Try increasing n (if possible)
if 24 < 24:  # This is always false, but included for completeness
    test_verification = verify_solution(25, 21)
    if test_verification['all_satisfied']:
        print(f"Can increase n to 25: p = {test_verification['objective_value']}")
    else:
        print("Increasing n violates constraints")
else:
    print("Cannot increase n beyond its upper bound")

# Try increasing r
test_verification = verify_solution(24, 22)
if test_verification['all_satisfied']:
    print(f"Can increase r to 22: p = {test_verification['objective_value']}")
else:
    print(f"Increasing r to 22 violates constraints: n*190 + r*310 = {24*190 + 22*310} > 11370")

print("\nConclusion: The optimal solution is n=24, r=21, p=330600")
```

```python
import numpy as np
from scipy.optimize import linprog

def solve_with_scipy_revised():
    """
    Minimize c = 0.21d + 0.29v
    Subject to:
    1. a >= 2520
    2. b >= 3150
    3. c >= 1260
    4. d = 40a + 50b + 20c
    5. v = 40a + 50b + 20c

    Using scipy.optimize.linprog
    """
    print("Solving the revised minimization problem using scipy.optimize.linprog")
    print("With constraints: d = 40a + 50b + 20c and v = 40a + 50b + 20c")

    # Step 1: Analyze the problem structure
    print("\nStep 1: Analyzing the problem structure")
    print("Since d = v = 40a + 50b + 20c, we can simplify the objective function:")
    print("c = 0.21d + 0.29v = 0.21(40a + 50b + 20c) + 0.29(40a + 50b + 20c)")
    print("c = (0.21 + 0.29)(40a + 50b + 20c) = 0.5(40a + 50b + 20c)")
    print("c = 20a + 25b + 10c")

    # Step 2: Define the problem in standard form
    print("\nStep 2: Defining the problem in standard form")

    # Decision variables: x = [a, b, c]
    # Objective function: minimize 20a + 25b + 10c
    c_obj = np.array([20, 25, 10])

    # Inequality constraints:
    # a >= 2520 -> -a <= -2520
    # b >= 3150 -> -b <= -3150
    # c >= 1260 -> -c <= -1260
    A_ub = np.array([
        [-1, 0, 0],  # -a <= -2520
        [0, -1, 0],  # -b <= -3150
        [0, 0, -1]   # -c <= -1260
    ])

    b_ub = np.array([-2520, -3150, -1260])

    print("Objective coefficients (c_obj):", c_obj)
    print("Inequality constraint matrix (A_ub):")
    print(A_ub)
    print("Inequality constraint bounds (b_ub):", b_ub)

    # Step 3: Solve the linear programming problem
    print("\nStep 3: Solving the linear programming problem")

    # All variables are non-negative by default in linprog
    result = linprog(c_obj, A_ub=A_ub, b_ub=b_ub, method='highs')

    print("Optimization status:", result.message)
    print("Number of iterations performed:", result.nit)

    # Step 4: Extract and interpret the results
    print("\nStep 4: Extracting and interpreting the results")

    if result.success:
        a, b, c = result.x

        # Calculate d and v
        d = 40*a + 50*b + 20*c
        v = 40*a + 50*b + 20*c

        # Calculate the original objective value
        original_obj = 0.21*d + 0.29*v

        print(f"a = {a}")
        print(f"b = {b}")
        print(f"c = {c}")
        print(f"d = 40a + 50b + 20c = {d}")
        print(f"v = 40a + 50b + 20c = {v}")
        print(f"Minimum cost (using simplified objective) = {result.fun}")
        print(f"Minimum cost (using original objective) = {original_obj}")

        # Verify the constraints
        print("\nVerifying constraints:")
        print(f"a >= 2520: {a} >= 2520 is {a >= 2520}")
        print(f"b >= 3150: {b} >= 3150 is {b >= 3150}")
        print(f"c >= 1260: {c} >= 1260 is {c >= 1260}")

        # Verify the equality constraints
        print(f"d = 40a + 50b + 20c: {d} = {40*a + 50*b + 20*c} is {np.isclose(d, 40*a + 50*b + 20*c)}")
        print(f"v = 40a + 50b + 20c: {v} = {40*a + 50*b + 20*c} is {np.isclose(v, 40*a + 50*b + 20*c)}")

        return {
            "a": a,
            "b": b,
            "c": c,
            "d": d,
            "v": v,
            "min_cost": original_obj,
            "success": result.success
        }
    else:
        print("Optimization failed:", result.message)
        return {"success": False}

def solve_analytically():
    """
    Solve the problem analytically
    """
    print("\nSolving the problem analytically:")

    # Since we're minimizing 20a + 25b + 10c and all coefficients are positive,
    # the minimum occurs at the lower bounds of the variables
    a = 2520
    b = 3150
    c = 1260

    # Calculate d and v
    d = 40*a + 50*b + 20*c
    v = 40*a + 50*b + 20*c

    # Calculate the objective value
    obj_value = 0.21*d + 0.29*v

    print(f"a = {a}")
    print(f"b = {b}")
    print(f"c = {c}")
    print(f"d = v = 40a + 50b + 20c = {d}")
    print(f"Minimum cost = 0.21d + 0.29v = {obj_value}")

    return {
        "a": a,
        "b": b,
        "c": c,
        "d": d,
        "v": v,
        "min_cost": obj_value
    }

# Solve the problem using scipy
scipy_solution = solve_with_scipy_revised()

# Solve the problem analytically
analytical_solution = solve_analytically()

# Compare the solutions
if scipy_solution["success"]:
    print("\nComparing solutions:")
    print(f"SciPy: a = {scipy_solution['a']}, b = {scipy_solution['b']}, c = {scipy_solution['c']}, cost = {scipy_solution['min_cost']}")
    print(f"Analytical: a = {analytical_solution['a']}, b = {analytical_solution['b']}, c = {analytical_solution['c']}, cost = {analytical_solution['min_cost']}")
    print(f"Difference in cost: {abs(scipy_solution['min_cost'] - analytical_solution['min_cost'])}")

# Explain the solution
print("\nExplanation of the solution:")
print("1. With the revised constraints d = 40a + 50b + 20c and v = 40a + 50b + 20c,")
print("   d and v are equal and both equal to 40a + 50b + 20c.")
print("2. This simplifies the objective function to:")
print("   c = 0.21d + 0.29v = 0.21(40a + 50b + 20c) + 0.29(40a + 50b + 20c)")
print("   c = (0.21 + 0.29)(40a + 50b + 20c) = 0.5(40a + 50b + 20c)")
print("   c = 20a + 25b + 10c")
print("3. Since all coefficients in the objective function are positive and we're minimizing,")
print("   the optimal solution is at the lower bounds of the variables:")
print(f"   a = {analytical_solution['a']}, b = {analytical_solution['b']}, c = {analytical_solution['c']}")
print(f"4. This gives d = v = {analytical_solution['d']}")
print(f"5. The minimum cost is {analytical_solution['min_cost']}")

# Calculate the total contribution of each variable to the objective
if scipy_solution["success"]:
    a, b, c = analytical_solution['a'], analytical_solution['b'], analytical_solution['c']
    total = 40*a + 50*b + 20*c

    print("\nContribution of each variable to the total 40a + 50b + 20c:")
    print(f"a: 40({a}) = {40*a} ({40*a/total*100:.2f}%)")
    print(f"b: 50({b}) = {50*b} ({50*b/total*100:.2f}%)")
    print(f"c: 20({c}) = {20*c} ({20*c/total*100:.2f}%)")

    print("\nIf we could reduce one of the lower bounds, which would be most beneficial?")
    print(f"Reducing a by 1 would reduce the cost by: {20}")
    print(f"Reducing b by 1 would reduce the cost by: {25}")
    print(f"Reducing c by 1 would reduce the cost by: {10}")
    print("Therefore, reducing b would have the largest impact on reducing the cost.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def solve_minimization_problem():
    """
    Minimize c = 325r + 460p
    Subject to:
    1. r <= 16 million
    2. 7 million <= p <= 10 million
    3. p + r >= 11 million
    4. r, p >= 0 (implicit non-negativity)
    """
    # Convert to millions for simplicity
    # Minimize c = 325r + 460p
    # Subject to:
    # 1. r <= 16
    # 2. p >= 7
    # 3. p <= 10
    # 4. p + r >= 11
    # 5. r >= 0
    # 6. p >= 0

    # Find the corner points of the feasible region
    # Corner points occur at the intersections of constraints

    # Corner point 1: Intersection of p = 7 and p + r = 11
    p1 = 7
    r1 = 11 - p1  # r1 = 4

    # Corner point 2: Intersection of p = 10 and p + r = 11
    p2 = 10
    r2 = 11 - p2  # r2 = 1

    # Corner point 3: Intersection of p = 10 and r = 16
    p3 = 10
    r3 = 16

    # Corner point 4: Intersection of p = 7 and r = 16
    p4 = 7
    r4 = 16

    # Calculate the objective value at each corner point
    c1 = 325 * r1 + 460 * p1
    c2 = 325 * r2 + 460 * p2
    c3 = 325 * r3 + 460 * p3
    c4 = 325 * r4 + 460 * p4

    # Create a list of corner points and their objective values
    corner_points = [
        {"p": p1, "r": r1, "c": c1},
        {"p": p2, "r": r2, "c": c2},
        {"p": p3, "r": r3, "c": c3},
        {"p": p4, "r": r4, "c": c4}
    ]

    # Find the corner point with the minimum objective value
    min_point = min(corner_points, key=lambda x: x["c"])

    return min_point

# Solve the problem
solution = solve_minimization_problem()

# Print the solution
print("Optimal Solution:")
print(f"p = {solution['p']} million")
print(f"r = {solution['r']} million")
print(f"Minimum cost c = {solution['c']} million = ${solution['c']:,}")

# Verify that the solution satisfies all constraints
p_opt = solution['p']
r_opt = solution['r']

print("\nVerifying constraints:")
print(f"1. r <= 16 million: {r_opt} <= 16 is {r_opt <= 16}")
print(f"2. p >= 7 million: {p_opt} >= 7 is {p_opt >= 7}")
print(f"3. p <= 10 million: {p_opt} <= 10 is {p_opt <= 10}")
print(f"4. p + r >= 11 million: {p_opt} + {r_opt} = {p_opt + r_opt} >= 11 is {p_opt + r_opt >= 11}")
print(f"5. r >= 0: {r_opt} >= 0 is {r_opt >= 0}")
print(f"6. p >= 0: {p_opt} >= 0 is {p_opt >= 0}")

# Explain why this is the optimal solution
print("\nExplanation of the optimal solution:")
print("1. Since we're minimizing c = 325r + 460p and both coefficients are positive,")
print("   we want to make r and p as small as possible within the constraints.")
print("2. The coefficient of p (460) is larger than the coefficient of r (325),")
print("   so we prefer to increase r rather than p when necessary.")
print("3. The constraint p >= 7 million forces p to be at least 7 million.")
print("4. The constraint p + r >= 11 million means that if p = 7, then r must be at least 4 million.")
print("5. This gives us the point (p = 7, r = 4), which satisfies all constraints.")
print("6. The minimum cost at this point is c = 325(4) + 460(7) = 4,520 million.")

# Visualize the feasible region and the optimal solution
plt.figure(figsize=(10, 8))

# Plot the feasible region
r = np.linspace(0, 20, 1000)
p1 = np.ones_like(r) * 7  # p >= 7
p2 = np.ones_like(r) * 10  # p <= 10
p3 = 11 - r  # p + r >= 11
r1 = np.ones_like(r) * 16  # r <= 16

plt.fill_between(r, p1, p2, where=(r <= 16) & (11 - r <= p2), alpha=0.3, color='blue')
plt.plot(r, p1, 'b-', label='p = 7')
plt.plot(r, p2, 'g-', label='p = 10')
plt.plot(r, p3, 'r-', label='p + r = 11')
plt.plot([16, 16], [0, 20], 'm-', label='r = 16')

# Plot the optimal solution
plt.plot(solution['r'], solution['p'], 'ro', markersize=10, label=f"Optimal: (r={solution['r']}, p={solution['p']})")

# Plot level curves of the objective function
c_values = [4000, 5000, 6000, 7000, 8000, 9000]
for c in c_values:
    p_level = (c - 325 * r) / 460
    plt.plot(r, p_level, 'k--', alpha=0.5)
    plt.text(15, (c - 325 * 15) / 460, f'c = {c}', fontsize=8)

plt.xlabel('r (millions)')
plt.ylabel('p (millions)')
plt.title('Minimization Problem: c = 325r + 460p')
plt.grid(True)
plt.legend()
plt.xlim(0, 20)
plt.ylim(0, 15)
plt.savefig('feasible_region.png')  # Save the plot as an image

print("\nA visualization of the feasible region and optimal solution has been created.")
```

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from scipy.misc import derivative

def numerical_approach():
    """
    Calculate the derivative using numerical approximation
    """
    print("Approach 1: Numerical Approximation")

    # Define the function
    def f(x):
        # Handle the case where x is exactly 7
        if np.isclose(x, 7):
            # Use the simplified form of the function
            return x**2 + 5*x - 5
        return (x**3 - 2*x**2 - 40*x + 35) / (x - 7)

    # Calculate the derivative numerically at points approaching 7
    h = 1e-5
    points = [6.9, 6.99, 6.999, 6.9999, 7.0001, 7.001, 7.01, 7.1]

    print("x values approaching 7 | Numerical derivative")
    print("------------------------|-----------------------")

    for x in points:
        # Use the central difference formula
        deriv = (f(x + h) - f(x - h)) / (2 * h)
        print(f"{x:.6f} | {deriv:.6f}")

    # Use scipy's derivative function
    result = derivative(f, 7, dx=1e-6, n=1)
    print(f"\nScipy's derivative at x=7: {result}")

    return result

def symbolic_approach():
    """
    Calculate the derivative using symbolic mathematics
    """
    print("\nApproach 2: Symbolic Mathematics with SymPy")

    # Define the symbolic variable and function
    x = sp.Symbol('x')
    f_expr = (x**3 - 2*x**2 - 40*x + 35) / (x - 7)

    # Simplify the function
    f_simplified = sp.simplify(f_expr)
    print(f"Simplified function: {f_simplified}")

    # Calculate the derivative
    f_prime = sp.diff(f_simplified, x)
    print(f"Derivative of simplified function: {f_prime}")

    # Evaluate at x = 7
    result = f_prime.subs(x, 7)
    print(f"Derivative evaluated at x=7: {result}")

    return result

def algebraic_approach():
    """
    Calculate the derivative using algebraic manipulation
    """
    print("\nApproach 3: Algebraic Manipulation")

    print("Step 1: Factor the numerator to simplify the function")
    print("(x^3-2x^2-40x+35)/(x-7)")

    print("\nCheck if the numerator equals 0 when x=7:")
    num_at_7 = 7**3 - 2*(7**2) - 40*7 + 35
    print(f"7^3-2(7^2)-40(7)+35 = {num_at_7}")

    print("\nSince the numerator equals 0 when x=7, we can factor (x-7) from the numerator:")
    print("x^3-2x^2-40x+35 = (x-7)(ax^2+bx+c)")

    print("\nExpanding the right side:")
    print("(x-7)(ax^2+bx+c) = ax^3+bx^2+cx-7ax^2-7bx-7c")

    print("\nComparing coefficients with the original expression:")
    print("x^3 term: a = 1")
    print("x^2 term: b-7a = -2, so b = -2+7 = 5")
    print("x^1 term: c-7b = -40, so c = -40+7(5) = -40+35 = -5")
    print("constant term: -7c = 35, so c = -5 (which matches our previous calculation)")

    print("\nTherefore:")
    print("x^3-2x^2-40x+35 = (x-7)(x^2+5x-5)")

    print("\nThis means our original function simplifies to:")
    print("f(x) = (x-7)(x^2+5x-5)/(x-7) = x^2+5x-5 for x ≠ 7")

    print("\nNow, to find the derivative of f(x) as x approaches 7, we can find the derivative of the simplified function x^2+5x-5 and evaluate it at x=7.")
    print("The derivative of x^2+5x-5 is 2x+5.")

    result = 2*7 + 5
    print(f"So the derivative of f(x) as x approaches 7 is 2(7)+5 = {result}")

    return result

def limit_approach():
    """
    Calculate the derivative using the limit definition
    """
    print("\nApproach 4: Limit Definition of Derivative")

    # Define the function
    def f(x):
        if np.isclose(x, 7):
            return 7**2 + 5*7 - 5  # Simplified form at x=7
        return (x**3 - 2*x**2 - 40*x + 35) / (x - 7)

    # Calculate the derivative using the limit definition
    h_values = [0.1, 0.01, 0.001, 0.0001, 0.00001]

    print("h values | Limit approximation")
    print("---------|--------------------")

    for h in h_values:
        limit_approx = (f(7 + h) - f(7)) / h
        print(f"{h:.5f} | {limit_approx:.6f}")

    # The limit as h approaches 0
    result = 19  # Based on our algebraic calculation
    print(f"\nAs h approaches 0, the limit approaches {result}")

    return result

def visualize_function_and_derivative():
    """
    Visualize the function and its derivative
    """
    print("\nVisualizing the function and its derivative")

    # Define the function and its derivative
    def f(x):
        # Use the simplified form to avoid division by zero
        return x**2 + 5*x - 5

    def f_prime(x):
        return 2*x + 5

    # Create x values
    x = np.linspace(4, 10, 1000)

    # Calculate function and derivative values
    y = f(x)
    y_prime = f_prime(x)

    # Create the plot
    plt.figure(figsize=(12, 8))

    # Plot the function
    plt.subplot(2, 1, 1)
    plt.plot(x, y, 'b-', label='f(x) = x²+5x-5')
    plt.axvline(x=7, color='r', linestyle='--', alpha=0.5, label='x = 7')
    plt.scatter([7], [f(7)], color='r', s=50, zorder=5)
    plt.annotate(f'f(7) = {f(7)}', xy=(7, f(7)), xytext=(7.2, f(7)+5),
                 arrowprops=dict(facecolor='black', shrink=0.05, width=1.5))
    plt.grid(True)
    plt.legend()
    plt.title('Function: f(x) = (x³-2x²-40x+35)/(x-7) = x²+5x-5')

    # Plot the derivative
    plt.subplot(2, 1, 2)
    plt.plot(x, y_prime, 'g-', label="f'(x) = 2x+5")
    plt.axvline(x=7, color='r', linestyle='--', alpha=0.5, label='x = 7')
    plt.scatter([7], [f_prime(7)], color='r', s=50, zorder=5)
    plt.annotate(f"f'(7) = {f_prime(7)}", xy=(7, f_prime(7)), xytext=(7.2, f_prime(7)+2),
                 arrowprops=dict(facecolor='black', shrink=0.05, width=1.5))
    plt.grid(True)
    plt.legend()
    plt.title('Derivative: f\'(x) = 2x+5')

    plt.tight_layout()
    plt.savefig('function_and_derivative.png')

    print("Visualization saved as 'function_and_derivative.png'")

# Execute all approaches
print("Finding the derivative of (x^3-2x^2-40x+35)/(x-7) as x approaches 7\n")

numerical_result = numerical_approach()
symbolic_result = symbolic_approach()
algebraic_result = algebraic_approach()
limit_result = limit_approach()

# Compare results
print("\nComparison of results:")
print(f"Numerical approach: {numerical_result}")
print(f"Symbolic approach: {symbolic_result}")
print(f"Algebraic approach: {algebraic_result}")
print(f"Limit approach: {limit_result}")

# Visualize the function and its derivative
visualize_function_and_derivative()

print("\nConclusion:")
print("The derivative of (x^3-2x^2-40x+35)/(x-7) as x approaches 7 is 19.")
```

# Find global and local maxima and minima

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and derivatives
def f(x):
    return 5 * (x - 5)**(2/3)

def f_prime(x):
    return 3.33333333333333/(x - 5)**0.333333333333333

def f_double_prime(x):
    return (-10/9)*(x-5)**(-4/3)

# Step 1: Find critical points (by solving f'(x) = 0 manually)
critical_points = np.array([5])  # from factoring 4x(x^2 -9)

# Step 2: Check endpoints
interval = [-100, 100]
endpoints = np.array(interval)

# Step 3: Combine points inside interval
all_points = np.concatenate((critical_points, endpoints))
all_points_in_interval = [x for x in all_points if interval[0] <= x <= interval[1]]

# Step 4: Evaluate function at all points
values = np.array([f(x) for x in all_points_in_interval])

# Step 5: Find global min and max
global_min_index = np.argmin(values)
global_max_index = np.argmax(values)
global_min_x = all_points_in_interval[global_min_index]
global_max_x = all_points_in_interval[global_max_index]
global_min_y = values[global_min_index]
global_max_y = values[global_max_index]

# Step 6: Get critical point nature
critical_info = []
for c in critical_points:
    if interval[0] <= c <= interval[1]:
        fpp = f_double_prime(c)
        nature = 'min' if fpp > 0 else 'max' if fpp < 0 else 'inconclusive'
        critical_info.append((c, f(c), nature))

# Step 7: Plot everything
x_vals = np.linspace(interval[0] - 1, interval[1] + 1, 400)
y_vals = f(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r'$f(x) = x^4 -18x^2 +11$')

# Mark critical points
for c, yc, nature in critical_info:
    color = 'red' if nature == 'max' else 'green' if nature == 'min' else 'blue'
    plt.plot(c, yc, 'o', color=color, markersize=10, label=f'{nature} at x={c}')
    plt.text(c, yc + 5, f'{nature}\nx={c:.1f}', ha='center')

# Mark endpoints
for e in endpoints:
    ey = f(e)
    plt.plot(e, ey, 's', color='purple', markersize=8, label=f'endpoint at x={e}')
    plt.text(e, ey + 5, f'endpoint\nx={e}', ha='center')

# Mark global extrema
plt.plot(global_min_x, global_min_y, '*', color='black', markersize=15, label='Global min')
plt.plot(global_max_x, global_max_y, '*', color='gold', markersize=15, label='Global max')
plt.text(global_min_x, global_min_y - 10, 'GLOBAL MIN', ha='center', color='black')
plt.text(global_max_x, global_max_y + 10, 'GLOBAL MAX', ha='center', color='gold')

plt.title('Graph of $f(x)$ with Critical Points, Endpoints, and Global Extrema')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

# Step 8: Print summary
print(f"Global minimum at x = {global_min_x} with value f(x) = {global_min_y}")
print(f"Global maximum at x = {global_max_x} with value f(x) = {global_max_y}")
```

```python
import sympy as sp

# Define symbol and function
x = sp.symbols('x')
f = 8 * (x - 4)**(2/3)
f_prime = sp.diff(f, x)

# Find where derivative is undefined
undefined_points = sp.solve(sp.denom(sp.simplify(f_prime)), x)
critical_points = sorted(undefined_points)

# Define explicit test points on each side of x = 5
test_points = [2, 6]

increasing_intervals = []
decreasing_intervals = []

for tp in test_points:
    derivative_value = f_prime.subs(x, tp).evalf()
    # Check if the derivative is real
    if derivative_value.is_real:
        slope = derivative_value
    else:
        # Take only the real part
        slope = sp.re(derivative_value)

    # Assign interval based on test point
    if tp < critical_points[0]:
        interval = (-sp.oo, critical_points[0])
    else:
        interval = (critical_points[0], sp.oo)

    # Check slope sign
    if slope > 0:
        increasing_intervals.append(interval)
    elif slope < 0:
        decreasing_intervals.append(interval)

# Print results
print(f"Derivative f'(x): {f_prime}")
print(f"Critical point at x = {critical_points[0]}")
print("Increasing intervals:", increasing_intervals)
print("Decreasing intervals:", decreasing_intervals)
```

# Newton's Method

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# Define the function and derivative
x = sp.symbols('x')
f_expr = x + 5 - sp.exp(x)
f_prime_expr = sp.diff(f_expr, x)

# Convert to numerical functions
f_func = sp.lambdify(x, f_expr, 'numpy')
f_prime_func = sp.lambdify(x, f_prime_expr, 'numpy')

# Newton's method parameters
x_n = 0.8  # initial guess
max_iter = 10
tolerance = 1e-6

# Set up plot range
x_vals = np.linspace(-1, 3, 400)
y_vals = f_func(x_vals)

# Store points for the list
iteration_points = []

# Plot function
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r'$f(x) = x + 5 - e^x$', color='black')
plt.axhline(0, color='gray', linestyle='--')

# Run Newton's method and plot tangent + steps
for i in range(max_iter):
    f_val = f_func(x_n)
    f_prime_val = f_prime_func(x_n)

    # Save current point
    iteration_points.append((i + 1, x_n, f_val))

    # Compute next point
    x_next = x_n - f_val / f_prime_val
    f_next = f_func(x_next)

    # Plot tangent line at x_n
    tangent_x = np.linspace(x_n - 1, x_n + 1, 10)
    tangent_y = f_val + f_prime_val * (tangent_x - x_n)
    plt.plot(tangent_x, tangent_y, '--', label=f'Iter {i + 1}')

    # Plot current point (on function)
    plt.plot(x_n, f_val, 'ro')  # red dot on curve

    # Plot vertical drop to x-axis for next guess
    plt.plot([x_next, x_next], [0, f_func(x_next)], ':b')  # vertical blue dotted line
    plt.plot(x_next, 0, 'bo')  # blue dot on x-axis

    # Check convergence
    if abs(x_next - x_n) < tolerance:
        break

    x_n = x_next

plt.title("Newton's Method with Tangent Steps (No Labels)")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

# Print summary table of points
print("\nIteration Summary:")
print(f"{'Iter':<5}{'x':>12}{'f(x)':>15}")
print("-" * 32)
for iter_num, x_val, f_val in iteration_points:
    print(f"{iter_num:<5}{x_val:>12.6f}{f_val:>15.6f}")
```

# Area between two curves

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from sympy import symbols, Eq, solve

# Step 1: Define functions
def f(x):
    return x**3 + 4*x**2 + 3*x + 2

def g(x):
    return x**3 + 3*x**2 + x + 17

# Step 2: Solve f(x) = g(x) for intersection points
x = symbols('x')
intersection_eq = Eq(x**3 + 4*x**2 + 3*x + 2, x**3 + 3*x**2 + x + 17)
solutions = solve(intersection_eq, x)
a, b = float(solutions[0]), float(solutions[1])

# Step 3: Define the absolute difference
def diff(x):
    return abs(f(x) - g(x))

# Step 4: Integrate the absolute difference
area, _ = quad(diff, a, b)
print(f"Intersection interval: [{a}, {b}]")
print(f"Enclosed area: {area:.6f} square units")

# Step 5: Plot the functions and shaded area
x_vals = np.linspace(a - 1, b + 1, 400)
plt.plot(x_vals, f(x_vals), label='f(x)', linewidth=2)
plt.plot(x_vals, g(x_vals), label='g(x)', linewidth=2)
plt.fill_between(x_vals, f(x_vals), g(x_vals), where=(f(x_vals) >= g(x_vals)),
                 interpolate=True, color='lightgray', label='Enclosed Area')
plt.fill_between(x_vals, f(x_vals), g(x_vals), where=(f(x_vals) < g(x_vals)),
                 interpolate=True, color='lightgray')
plt.title("Graph of f(x) and g(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# ---------------------
# Symbolic Setup
# ---------------------
x = sp.Symbol('x', real=True)

# Define the function
f_expr = sp.sqrt(2 * x**3 + 9 * x**2)

# Domain: where the expression inside sqrt is >= 0
radicand = 2 * x**3 + 9 * x**2
domain_condition = sp.solve(radicand >= 0, x)

# Derivative
f_prime = sp.diff(f_expr, x)

# Critical points: where f'(x) = 0
critical_points = sp.solve(f_prime, x)
critical_points = [p.evalf() for p in critical_points if p.is_real]

# Evaluate function at critical points
extrema = [(float(p), float(f_expr.subs(x, p).evalf())) for p in critical_points]

# Second derivative test to classify extrema
f_double_prime = sp.diff(f_prime, x)
min_points = []
max_points = []

for pt in critical_points:
    second_deriv = f_double_prime.subs(x, pt).evalf()
    y_val = float(f_expr.subs(x, pt).evalf())
    if second_deriv > 0:
        min_points.append((float(pt), y_val))
    elif second_deriv < 0:
        max_points.append((float(pt), y_val))

# ---------------------
# Numeric Evaluation for Plot
# ---------------------
x_vals = np.linspace(-4.5, 5, 1000)
x_vals = x_vals[2:]  # Avoid invalid sqrt region near domain edge
y_vals = np.sqrt(2 * x_vals**3 + 9 * x_vals**2)

# ---------------------
# Plot the function and extrema
# ---------------------
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r'$f(x) = \sqrt{2x^3 + 9x^2}$', color='blue')

# Mark local min
for x_min, y_min in min_points:
    plt.plot(x_min, y_min, 'go', label=f'Local Min: x={x_min:.2f}, f(x)={y_min:.2f}')

# Mark local max
for x_max, y_max in max_points:
    plt.plot(x_max, y_max, 'ro', label=f'Local Max: x={x_max:.2f}, f(x)={y_max:.2f}')

plt.title("Maxima and Minima of $f(x) = \\sqrt{2x^3 + 9x^2}$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.ylim(0, max(y_vals) + 10)
plt.xlim(-5, 5)
plt.show()

# ---------------------
# Output Results
# ---------------------
print("DOMAIN: x ≥ -9/2")
print("\nLocal Maximum Points:")
for xc, yc in max_points:
    print(f"  x = {xc:.2f}, f(x) = {yc:.2f}")

print("\nLocal Minimum Points:")
for xc, yc in min_points:
    print(f"  x = {xc:.2f}, f(x) = {yc:.2f}")
```

```python
import sympy as sp

# Define symbolic variable and function
x = sp.Symbol('x', real=True)
f = sp.sqrt(2 * x**3 + 9 * x**2)

# Compute derivative
f_prime = sp.diff(f, x)

# 1. Find where f'(x) = 0
critical_eq = sp.Eq(f_prime, 0)
crit_points_zero = sp.solve(critical_eq, x)
crit_points_zero = [p.evalf() for p in crit_points_zero if p.is_real]

# 2. Find where f'(x) is undefined: where sqrt denominator = 0
# From the derivative's structure, the denominator is sqrt(2x^3 + 9x^2)
undefined_points = sp.solve(sp.Eq(2 * x**3 + 9 * x**2, 0), x)
undefined_points = [p.evalf() for p in undefined_points if p.is_real]

# 3. Combine and remove duplicates
all_critical_points = sorted(set(crit_points_zero + undefined_points))

# 4. Evaluate f(x) at each critical point
print("Critical Points (x, f(x)):")
for pt in all_critical_points:
    f_val = f.subs(x, pt).evalf()
    print(f"  x = {pt:.4f}, f(x) = {f_val:.4f}")
```

```python
import math

# Define the function and its derivative
def f(x):
    return 7 * math.sin(x) - x

def f_prime(x):
    return 7 * math.cos(x) - 1

# Initial guess
x_n = 2

# Iteration settings
tolerance = 1e-6
max_iterations = 100

print("Newton's Method for solving 7sin(x) = x\n")
for i in range(max_iterations):
    f_xn = f(x_n)
    f_prime_xn = f_prime(x_n)

    if f_prime_xn == 0:
        print("Zero derivative. No solution found.")
        break

    x_next = x_n - f_xn / f_prime_xn

    print(f"Iteration {i+1}: x = {x_next:.6f}")

    if abs(x_next - x_n) < tolerance:
        print(f"\nConverged to solution x ≈ {x_next:.6f} after {i+1} iterations.")
        break

    x_n = x_next
else:
    print("Did not converge within maximum iterations.")
```

```python
import sympy as sp
from ipywidgets import interact, Textarea, Output
from IPython.display import display

# Define symbolic variables
x, y = sp.symbols('x y')

# Output area
out = Output()

def analyze_critical_points(expr_str):
    out.clear_output()
    try:
        # Parse the function
        f = sp.sympify(expr_str)

        # First partial derivatives
        fx = sp.diff(f, x)
        fy = sp.diff(f, y)

        # Solve the system fx = 0 and fy = 0
        critical_points = sp.solve([fx, fy], (x, y), dict=True)

        with out:
            print("Function:")
            display(f)
            print("\nFirst Partial Derivatives:")
            print("∂f/∂x =")
            display(fx)
            print("∂f/∂y =")
            display(fy)

            print("\nCritical Points:")
            if not critical_points:
                print("No critical points found.")
            else:
                for pt in critical_points:
                    display(pt)

            # Classify critical points using second derivative test
            fxx = sp.diff(fx, x)
            fyy = sp.diff(fy, y)
            fxy = sp.diff(fx, y)

            print("\nSecond Partial Derivatives:")
            print("f_xx ="), display(fxx)
            print("f_yy ="), display(fyy)
            print("f_xy ="), display(fxy)

            for pt in critical_points:
                H = fxx.subs(pt) * fyy.subs(pt) - fxy.subs(pt)**2
                print(f"\nAt point {pt}:")
                print(f"Determinant of Hessian (D) = {H}")
                if H > 0:
                    if fxx.subs(pt) > 0:
                        print("→ Local Minimum")
                    elif fxx.subs(pt) < 0:
                        print("→ Local Maximum")
                    else:
                        print("→ Inconclusive (f_xx = 0)")
                elif H < 0:
                    print("→ Saddle Point")
                else:
                    print("→ Inconclusive (Hessian determinant = 0)")
    except Exception as e:
        with out:
            print(f"Error parsing input: {e}")

# Input widget
input_box = Textarea(
    value='x**2 + y**2 + x*y',
    description='f(x, y):',
    layout={'width': '100%', 'height': '80px'}
)

interact(analyze_critical_points, expr_str=input_box)
display(out)
```

```python
import sympy as sp

# === Define symbolic variables ===
p, r = sp.symbols('p r', positive=True, real=True)

# === Define parameters (you can edit these) ===
A = 20              # coefficient
a = 0.75                # exponent on p
b = 0.25              # exponent on r

c1 = 300             # cost per unit p
c2 = 2100            # cost per unit r
budget = 336000      # total available cost

# === Define production function and constraint ===
z = A * p**a * r**b
constraint = sp.Eq(c1*p + c2*r, budget)

# === Solve the constraint for p in terms of r ===
p_sub = sp.solve(constraint, p)[0]

# === Substitute into z ===
z_sub = z.subs(p, p_sub)

# === Differentiate z(r) and find maximum ===
dz_dr = sp.diff(z_sub, r)
critical_points = sp.solve(dz_dr, r)

# === Filter valid real solution ===
r_star = [pt.evalf() for pt in critical_points if pt.is_real and pt > 0][0]
p_star = p_sub.subs(r, r_star).evalf()
z_star = z.subs({p: p_star, r: r_star}).evalf()

# === Print results ===
print(f"Optimal p: {p_star}")
print(f"Optimal r: {r_star}")
print(f"Maximum z: {z_star}")
```

```python
import sympy as sp
from sympy import symbols
from ipywidgets import interact, Textarea, Output
from IPython.display import display

# Define symbols
x, y = sp.symbols('x y', real=True)

# Output area for results
out = Output()

def analyze_function(expr_str):
    out.clear_output()
    try:
        # Preprocess: replace ^ with ** to ensure valid Python exponentiation
        expr_str = expr_str.replace('^', '**')

        # Safely parse the expression
        f = sp.sympify(expr_str, evaluate=False)

        # First partial derivatives
        fx = sp.diff(f, x)
        fy = sp.diff(f, y)

        # Solve for critical points
        solutions = sp.nonlinsolve([fx, fy], (x, y))
        crit_points = [dict(zip((x, y), sol)) for sol in solutions if all(val.is_real for val in sol)]

        # Second derivatives
        fxx = sp.diff(fx, x)
        fyy = sp.diff(fy, y)
        fxy = sp.diff(fx, y)

        with out:
            print(f"Function f(x, y) = {f}")
            print("\nFirst Partial Derivatives:")
            print("∂f/∂x =", fx)
            print("∂f/∂y =", fy)

            print("\nCritical Points and Classification:")
            if not crit_points:
                print("No critical points found.")
            else:
                for pt in crit_points:
                    x0, y0 = pt[x], pt[y]
                    fx_val = fxx.subs(pt)
                    fy_val = fyy.subs(pt)
                    fxy_val = fxy.subs(pt)
                    D = fx_val * fy_val - fxy_val**2

                    print(f"\nPoint: x = {x0}, y = {y0}")
                    print(f"f_xx = {fx_val}, f_yy = {fy_val}, f_xy = {fxy_val}")
                    print(f"Hessian determinant D = {D}")

                    if D > 0:
                        if fx_val > 0:
                            print("→ Local Minimum")
                        elif fx_val < 0:
                            print("→ Local Maximum")
                        else:
                            print("→ Inconclusive (f_xx = 0)")
                    elif D < 0:
                        print("→ Saddle Point")
                    else:
                        print("→ Inconclusive (D = 0)")
    except Exception as e:
        with out:
            print(f"Error: {e}")


# Text input for user-defined function
input_box = Textarea(
    value='4*x**4 + 4*y**4 - 2*x*y',
    description='f(x, y):',
    layout={'width': '100%', 'height': '80px'}
)

interact(analyze_function, expr_str=input_box)
display(out)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar
from math import pi

# Fixed volume
V = 432*pi

# Surface area function as a function of radius r
def surface_area(r):
    h = V / (pi * r**2)  # Height from volume constraint
    return 2 * pi * r**2 + 2 * pi * r * h  # Total surface area

# Minimize surface area with respect to radius
res = minimize_scalar(surface_area, bounds=(0.1, 10), method='bounded')
r_opt = res.x
h_opt = V / (pi * r_opt**2)
sa_opt = surface_area(r_opt)

# Print results
print(f"Optimal radius (r): {r_opt:.2f}")
print(f"Optimal height (h): {h_opt:.2f}")
print(f"Minimum surface area: {sa_opt:.2f}")

# Visualization
r_vals = np.linspace(0.1, 5, 300)
sa_vals = [surface_area(r) for r in r_vals]

plt.figure(figsize=(10, 6))
plt.plot(r_vals, sa_vals, label="Surface Area")
plt.axvline(r_opt, color='red', linestyle='--', label=f"Optimal r ≈ {r_opt:.2f}")
plt.title("Surface Area of a Cylinder vs Radius (Fixed Volume = 54)")
plt.xlabel("Radius")
plt.ylabel("Surface Area")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from sympy import symbols, Eq, solve

# Step 1: Define functions
def f(x):
    return 5 / x

def g(x):
    return 7 - x

# Step 2: Solve f(x) = g(x) for intersection points
x = symbols('x')
intersection_eq = Eq(5 / x, 7 - x)
solutions = solve(intersection_eq, x)

# Filter real and positive solutions
real_solutions = [sol.evalf() for sol in solutions if sol.is_real and sol.evalf() > 0]
a, b = float(min(real_solutions)), float(max(real_solutions))

# Step 3: Define the absolute difference
def diff(x_val):
    return abs(f(x_val) - g(x_val))

# Step 4: Integrate the absolute difference
area, _ = quad(diff, a, b)
print(f"Intersection interval: [{a:.3f}, {b:.3f}]")
print(f"Enclosed area: {area:.6f} square units")

# Step 5: Plot the functions and shaded area
x_vals = np.linspace(a - 1, b + 1, 400)
f_vals = f(x_vals)
g_vals = g(x_vals)

plt.plot(x_vals, f_vals, label='f(x) = 5/x', linewidth=2)
plt.plot(x_vals, g_vals, label='g(x) = 7 - x', linewidth=2)
plt.fill_between(x_vals, f_vals, g_vals, where=(f_vals >= g_vals),
                 interpolate=True, color='lightgray', label='Enclosed Area')
plt.fill_between(x_vals, f_vals, g_vals, where=(f_vals < g_vals),
                 interpolate=True, color='lightgray')
plt.title("Graph of f(x) and g(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(bottom=0, top=10)
plt.legend()
plt.grid(True)
plt.show()
```
